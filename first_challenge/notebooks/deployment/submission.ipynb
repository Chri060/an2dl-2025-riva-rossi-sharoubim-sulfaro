## Model definition

```python
# Parameters to define the name of the model to be submitted
MODEL = 'efficientnetv2b3_again_89.26'
```

```python
# Creation of a file called model.py
%%writefile model.py
import numpy as np
import tensorflow as tf
tfk = tf.keras
tfkl = tfk.layers

# Definition of the class for the model
class Model:
    # Constructor of the Model class
    def __init__(self):
        self.neural_network = tfk.models.load_model(f'weights.keras')

    # Method used to predict the labels of a images' set
    def predict(self, X):
        preds = self.neural_network.predict(X)
        if len(preds.shape) == 2:
            preds = np.argmax(preds, axis=1)
        return preds
```
